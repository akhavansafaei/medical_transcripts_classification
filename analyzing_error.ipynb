```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming you have already renamed the "medical_specialty" column to "label"

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Fit the encoder to your unique labels and transform the "label" column

# Load the train and test CSV files
train = pd.read_csv('/content/drive/My Drive/training.csv')
valid=pd.read_csv('/content/drive/My Drive/validation.csv')
test = pd.read_csv('/content/drive/My Drive/test.csv')
# List of categories
categories = [
    "Surgery",
    "Consult - History and Phy.",
    "Cardiovascular / Pulmonary",
    "Orthopedic",
    "Radiology",
    "General Medicine",
    "Gastroenterology",
    "Neurology",
    "SOAP / Chart / Progress Notes",
    "Obstetrics / Gynecology",
    "Urology",
    "Discharge Summary"
]
specialty_numbers=[38, 5, 3, 27, 33, 15, 14, 22, 35, 24, 39, 10]

# Iterate through the 'label' column of train, test, and valid dataframes
for df in [train, test, valid]:
    df['label'] = df['label'].apply(lambda x: x if x in specialty_numbers else 40)
    df['label'] = label_encoder.fit_transform(df['label'])

train=train[['text','ptext','index','label']]
valid=valid[['text','ptext','index','label']]
test=test[['text','ptext','index','label']]
# Extract features (X) and labels (y) for train and test datasets
X_train = train[['text', 'ptext']]
y_train = train['label']

X_valid=valid[['text', 'ptext']]
y_valid = valid['label']

X_test = test[['text', 'ptext']]
y_test = test['label']
```

```python
import zipfile

# Specify the path to the zip file
zip_file_path = '/content/drive/MyDrive/results_r.zip'

# Specify the extraction path
extraction_path = '/content/'

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_path)

print(f"File '{zip_file_path}' has been successfully unzipped to '{extraction_path}'.")
```

```python
import os
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix

# Specify the directory containing CSV files
directory = '/content/results_r'

# List all CSV files in the directory
csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]

# Loop through each CSV file
for csv_file in csv_files:
    # Read the CSV file
    df = pd.read_csv(os.path.join(directory, csv_file))

    # Assuming the CSV file contains a column named 'predictions' with model predictions
    model_predictions = df['predictions']  # Adjust the column name as needed

    # Assuming y_test is the ground truth labels
    # Calculate classification report and other error analysis metrics
    classification_rep = classification_report(y_test, model_predictions)
    confusion_mat = confusion_matrix(y_test, model_predictions)

    # Print the unique name of the CSV file and its corresponding analysis
    print(f"Analysis for '{csv_file}':")
    print("Classification Report:")
    print(classification_rep)
    print("Confusion Matrix:")
    print(confusion_mat)
    print("\n")
```
